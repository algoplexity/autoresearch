## Runtime Checklist

Before starting a longer run, confirm the following:
- GPU type is known (`T4`, `L4`, `A100`, etc.)
- CUDA is available in PyTorch
- package installs completed cleanly
- enough RAM / VRAM is available for the planned batch size
- flash attention imports resolve if required by the training path


```python
import os
import platform
import torch

print('Platform:', platform.platform())
print('Python:', platform.python_version())
print('CUDA available:', torch.cuda.is_available())
if torch.cuda.is_available():
    print('GPU:', torch.cuda.get_device_name(0))
    print('Capability:', torch.cuda.get_device_capability(0))
    props = torch.cuda.get_device_properties(0)
    print('Total VRAM (GB):', round(props.total_memory / 1024**3, 2))
```

## Known-good baseline config

Start with conservative defaults before scaling. Example baseline:
- `dtype = bfloat16` on newer GPUs when stable
- `dtype = float16` fallback on T4
- reduced `DEVICE_BATCH_SIZE`
- short run length for initial validation
- small shard count for first data prep checks


```python
baseline = {
    'dtype_preference': 'bfloat16 if supported, else float16',
    'device_batch_size': 16,
    'prep_num_shards': 4,
    'purpose': 'validate stability before scaling'
}
baseline
```

## Scaling Plan

Scale in stages instead of changing everything at once:
1. confirm data prep works on a small run
2. confirm training starts and completes a short smoke test
3. increase training length
4. increase shard count or dataset coverage
5. increase batch size only after memory headroom is clear


## Failure Patterns

Common Colab failure modes to watch for:
- out-of-memory during forward pass
- out-of-memory during backward pass
- dtype incompatibility on older GPUs
- package mismatch after runtime restart
- missing kernel/helper imports from upstream assumptions
- silent notebook reset after long idle periods
